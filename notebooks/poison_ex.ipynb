## Suppose we have a dataset where 
x represents the number of hours studi , an
ï¿½
y represents the exam score. In a typical scenario, we expect a positive correlation between hours studied and exam score.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate a clean dataset
np.random.seed(0)  # For reproducibility
x_clean = np.random.normal(5, 2, 30)  # Hours studied
y_clean = x_clean * 10 + np.random.normal(0, 5, 30)  # Exam scores

# Visualize the clean dataset
plt.scatter(x_clean, y_clean, label='Clean Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Clean Dataset')
plt.legend()
plt.show()
```

```python
# Adding poisoned data points
x_poison = np.array([8, 9, 10])  # High hours studied
y_poison = np.array([10, 12, 15])  # Anomalously low exam scores

# Combine the clean and poisoned data
x_combined = np.concatenate([x_clean, x_poison])
y_combined = np.concatenate([y_clean, y_poison])

# Visualize the dataset with poisoned data
plt.scatter(x_clean, y_clean, label='Clean Data')
plt.scatter(x_poison, y_poison, color='red', label='Poisoned Data')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Dataset with Poisoned Data')
plt.legend()
plt.show()
```

```python
# Train on clean data
model_clean = LinearRegression().fit(x_clean.reshape(-1, 1), y_clean)

# Train on combined data
model_poisoned = LinearRegression().fit(x_combined.reshape(-1, 1), y_combined)

# Visualize the impact of poisoning
plt.scatter(x_combined, y_combined, label='Data Points')
plt.plot(x_combined, model_clean.predict(x_combined.reshape(-1, 1)), label='Clean Model', color='green')
plt.plot(x_combined, model_poisoned.predict(x_combined.reshape(-1, 1)), label='Poisoned Model', color='red')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Impact of Data Poisoning on Linear Regression')
plt.legend()
plt.show()
```
